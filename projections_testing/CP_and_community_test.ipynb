## Community checks

```python
# For Bipartite Graph
#bipartite_communities = list(community.louvain_communities(G_bipartite))

# For Simple Projection
simple_projection_communities = community.louvain_communities(simple_proj_graph,weight = 'weight')


# For Allocation Projection
allocation_projection_communities = community.louvain_communities(allow_proj_graph,weight='weight')
```

```python
# For Simple Projection
simple_projection_communities = community.louvain_communities(simple_proj_graph, weight='weight')

# For Allocation Projection
allocation_projection_communities = community.louvain_communities(allow_proj_graph, weight='weight')

def communities_to_dict(communities):
    community_dict = {}
    for i, community_set in enumerate(communities):
        community_dict[i] = list(community_set)
    return community_dict

# Convert Louvain communities to dictionaries
simple_projection_dict = communities_to_dict(simple_projection_communities)
allocation_projection_dict = communities_to_dict(allocation_projection_communities)

# Print or use the dictionaries as needed
#print("Simple Projection Communities:", simple_projection_dict)
#print("Allocation Projection Communities:", allocation_projection_dict)
```

```python


# Create a graph from the Simple Projection
G_simple = nx.Graph(simple_proj_graph)

# Assign a color to each community
community_colors = {}
for i, community_set in enumerate(simple_projection_communities):
    for node in community_set:
        community_colors[node] = i

# Generate a list of colors for each node based on its community
node_colors = [community_colors[node] for node in G_simple.nodes]

# Draw the graph with nodes colored by community
pos = nx.spring_layout(G_simple)  # You can use a different layout if needed
nx.draw(G_simple, pos, node_color=node_colors, cmap=plt.cm.jet, with_labels=False)
plt.show()

```

```python
# Create a graph from the Allowcation Projection
G_simple = nx.Graph(allow_proj_graph)

# Assign a color to each community
community_colors = {}
for i, community_set in enumerate(allocation_projection_communities):
    for node in community_set:
        community_colors[node] = i

# Generate a list of colors for each node based on its community
node_colors = [community_colors[node] for node in G_simple.nodes]

# Draw the graph with nodes colored by community
pos = nx.spring_layout(G_simple)  # You can use a different layout if needed
nx.draw(G_simple, pos, node_color=node_colors, cmap=plt.cm.jet, with_labels=False)
plt.show()
```
