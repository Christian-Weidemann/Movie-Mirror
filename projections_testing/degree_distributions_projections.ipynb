```python
import pickle
import matplotlib.pyplot as plt  
import networkx as nx
from networkx.algorithms import community
import csv
import pandas as pd
```

```python
with open('simple_weights_movies.p', 'rb') as file:
    simple = pickle.load(file)
    
with open('rating_allocation_movies.p', 'rb') as file:
    allow = pickle.load(file)

with open('rating_genre_allocation_movies.p', 'rb') as file:
    genre_allow = pickle.load(file)
```

```python
simple_proj_graph = nx.Graph(simple)
allow_proj_graph = nx.DiGraph(allow)
genre_allow_proj_graph = nx.DiGraph(genre_allow)
```

## For simple weights projection

```python
# Calculate the degree distribution
degrees = [simple_proj_graph.degree(node) for node in simple_proj_graph.nodes()]

# Count the occurrences of each degree
degree_counts = {degree: degrees.count(degree) for degree in set(degrees)}

# Extract the degrees and counts for plotting
degree_values = list(degree_counts.keys())
count_values = list(degree_counts.values())

# Plot the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(degree_values, count_values, marker='o', color='orange')
plt.title('Degree Distribution of Simple Weights Projection ')
plt.xlabel('Degree')
plt.ylabel('Count')
#plt.show()
#plt.close()
plt.savefig('degree_distribution_simple_weights_plot.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming simple_proj_graph is your networkx graph

# Calculate the weighted degree distribution
weighted_degrees = [simple_proj_graph.degree(node, weight='weight') for node in simple_proj_graph.nodes()]

# Count the occurrences of each weighted degree
weighted_degree_counts = {degree: weighted_degrees.count(degree) for degree in set(weighted_degrees)}

# Extract the weighted degrees and counts for plotting
weighted_degree_values = list(weighted_degree_counts.keys())
count_values = list(weighted_degree_counts.values())

# Sort weighted degree values
sorted_weighted_degrees = np.sort(weighted_degree_values)

# Calculate cumulative sum
cumulative_sum = np.cumsum(sorted_weighted_degrees)

# Calculate cumulative probability
cumulative_probability = cumulative_sum / np.sum(sorted_weighted_degrees)

# Plot the cumulative probability of weighted degree distribution
plt.figure(figsize=(8, 5))
plt.plot(sorted_weighted_degrees, cumulative_probability, marker='o', color='orange')
plt.title('Cumulative Probability of Weighted Degree Distribution of Simple Weights Projection')
plt.xlabel('Weighted Degree')
plt.ylabel('Cumulative Probability')
plt.gca().invert_yaxis()  # Invert y-axis to have higher values at the top
#plt.show()
plt.savefig('cumulative_prob_of_weighted_degree_distribution_simple_weights_plot.png')
```
