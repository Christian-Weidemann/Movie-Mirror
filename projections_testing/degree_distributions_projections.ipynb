```python
import pickle
import matplotlib.pyplot as plt  
import networkx as nx
from networkx.algorithms import community
import csv
import pandas as pd
```

```python
with open('simple_weights_movies.p', 'rb') as file:
    simple = pickle.load(file)
    
with open('rating_allocation_movies.p', 'rb') as file:
    allow = pickle.load(file)
```

```python
simple_proj_graph = nx.Graph(simple)
allow_proj_graph = nx.DiGraph(allow)
```

## For simple weights projection

```python
# Calculate the degree distribution
degrees = [simple_proj_graph.degree(node) for node in simple_proj_graph.nodes()]

# Count the occurrences of each degree
degree_counts = {degree: degrees.count(degree) for degree in set(degrees)}

# Extract the degrees and counts for plotting
degree_values = list(degree_counts.keys())
count_values = list(degree_counts.values())

# Plot the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(degree_values, count_values, marker='o', color='orange')
plt.title('Degree Distribution of Simple Weights Projection ')
plt.xlabel('Degree')
plt.ylabel('Count')
#plt.show()
#plt.close()
plt.savefig('degree_distribution_simple_weights_plot.png')
```

## For Weighted Resource Allocation

```python
# Calculate the in-weight degree distribution
in_degrees = dict(allow_proj_graph.in_degree(weight='weight'))

# Calculate the average in-weight degree
average_in_degree = sum(in_degrees.values()) / len(in_degrees)

# Count the occurrences of each in-degree
in_degree_counts = {in_degree: list(in_degrees.values()).count(in_degree) for in_degree in set(in_degrees.values())}

# Extract the in-degrees and counts for plotting
in_degree_values = list(in_degree_counts.keys())
count_values = list(in_degree_counts.values())

# Set the size of the figure
plt.figure(figsize=(8, 5))

# Plot the scatter plot
plt.scatter(in_degree_values, count_values, marker='o', color='orange')
plt.title('Average In-weight Degree of Weighted Resource Allocation Projection')
plt.xlabel('In-weight Degree')
plt.ylabel('Count')
#plt.show()
#plt.close()
plt.savefig('avg_indegree_distribution_plot_r_allocation.png')
```
