```python
import pickle
import matplotlib.pyplot as plt  
import networkx as nx

```

```python
with open('simple_weights_movies.p', 'rb') as file:
    simple = pickle.load(file)
```

```python
with open('rating_allocation_movies.p', 'rb') as file:
    allow = pickle.load(file)
```

```python
#simple_proj_graph = nx.read_gpickle(loaded_graph_data)
simple_proj_graph = nx.Graph(simple)
allow_proj_graph = nx.Graph(allow)
```

```python
# Compute the periphery for the simple_proj_graph
simple_periphery = nx.periphery(simple_proj_graph)

# Compute the periphery for the allow_proj_graph
allow_periphery = nx.periphery(allow_proj_graph)

# Print or analyze the results
print("Periphery of simple_proj_graph:", simple_periphery)
print("Periphery of allow_proj_graph:", allow_periphery)
```

```python
# Calculate the core number for each vertex
core_numbers_simple = nx.core_number(simple_proj_graph)
print("Core Numbers simple:", core_numbers_simple)

# Calculate the core number for each vertex
core_numbers_allow = nx.core_number(allow_proj_graph)
print("Core Numbers allow:", core_numbers_allow)

# Get the 2-core of the graph
k_core_2_simple = nx.k_core(simple_proj_graph, k=2)
print("2-Core simple projection:", k_core_2_simple)


k_core_2_allow = nx.k_core(allow_proj_graph, k=2)
print("2-Core allowcation:", k_core_2_allow)


# Get the 2-shell of the graph
k_shell_2_simple = nx.k_shell(simple_proj_graph, k=2)
print("2-Shell simple:", k_shell_2_simple)

# Get the 2-shell of the graph
k_shell_2_allow = nx.k_shell(allow_proj_graph, k=2)
print("2-Shell: allow ", k_shell_2_allow)
```
